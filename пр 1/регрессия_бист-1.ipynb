```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(15,8)})

import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn.model_selection import train_test_split
```

# Знакомство с данными

```python
data= pd.read_csv('AB_NYC_2019.csv')
data.info()
```

```python
data.head()
```

```python
data.drop('last_review', axis =1, inplace = True)
```

```python
data.last_review = pd.to_datetime(data.last_review)
data.info()
```

 id - уникальный идентификатор (с ресурса)
 name - название объявление 
 host_id - идентификатор хоста
 host_name - имя хоста
 neighbourhood_group - округ
 neighbourhood - район
     latitude и longitude - координаты квартиры
    room_type - тип сдаваемого помещения
    price - цена за помещение за сутки
    minimum_nights - минимальное количечство ночей на которое сдается помещение
    number_of_reviews - количество отзывов
    last_review - время последнего отзыва
    reviews_per_month - количество отзывов в месяц
    calculated_host_listings_count - количество объявлений хоста
    availability_365 - количество дней в году, когда объект доступен для бронирования

```python
data.drop(['id','name','host_id','host_name'],axis = 1, inplace = True)
data.drop_duplicates(inplace = True)
data.sort_values(by='last_review',inplace = True)
data.number_of_reviews.fillna(0.0, inplace = True)
data.info()
```

```python
data.reviews_per_month = data.reviews_per_month.astype('str')
data.reviews_per_month = data.reviews_per_month.apply(lambda x: '0.00' if x == 'nan' else x)
data.reviews_per_month = data.reviews_per_month.astype(np.float64)
data.info()
```

# Визуализация и анализ данных

```python
room_pivot = pd.pivot_table(data,'price',['neighbourhood_group'],aggfunc=np.mean)
room_pivot
```

```python
agg_price = data.groupby('neighbourhood_group').agg({'price':'mean','number_of_reviews':'mean'}).reset_index()
agg_price
```

```python
sns.barplot(data=agg_price,x='neighbourhood_group',y='price')
```

```python
g=data[data.price <500]
plt.figure(figsize=(10,6))
sns.boxplot(y="price",x="neighbourhood_group",data=g)
plt.title("neighbourhood_group price disturbtion <500");
```

```python
map_ng_to_uniq_cnt = {}
for row in data.groupby('neighbourhood_group').agg({'neighbourhood':'nunique'}).reset_index().values:
    map_ng_to_uniq_cnt[row[0]]=row[1]

data['uniq_neighbourhood_in_group']=data['neighbourhood_group'].map(map_ng_to_uniq_cnt)
data.head()
```

```python
agg_price_by_roomtype = data.groupby('room_type').agg({'price':'mean'}).reset_index()
agg_price_by_roomtype
```

```python
agg_price_by_roomtype_and_nggroup = data.groupby(['neighbourhood_group','room_type']).agg({'price':'mean'}).reset_index()
agg_price_by_roomtype_and_nggroup
```

```python
sns.catplot(x='room_type',y='price',hue='neighbourhood_group',kind='point',data=agg_price_by_roomtype_and_nggroup)
```
