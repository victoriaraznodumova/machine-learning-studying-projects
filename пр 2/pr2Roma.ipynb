```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("credit_score_prediction.csv")
```

```python
data = data.drop(["ID", "Customer_ID", "SSN"], axis=True)
```

```python
import re
def clean_number(value):
    if pd.isna(value):
        return None
    value = str(value)
    value = re.sub(r'[^\d,.]', '', value) 
    return float(value) if value else None
columns = ["Annual_Income","Age", "Monthly_Inhand_Salary", "Num_Bank_Accounts", "Num_Credit_Card", "Interest_Rate", "Num_of_Loan", "Delay_from_due_date", "Num_of_Delayed_Payment", "Changed_Credit_Limit", "Num_Credit_Inquiries", "Outstanding_Debt", "Credit_Utilization_Ratio", "Total_EMI_per_month", "Amount_invested_monthly", "Monthly_Balance"]
data[columns] = data[columns].applymap(clean_number)

data.head()
```

```python
import re
def clean_text(value):
    if pd.isna(value): 
        return ""
    value = str(value).strip() 
    value = re.sub(r'[^a-zA-Zа-яА-Я0-9_\s]', '', value)  
    value = value.replace('_', ' ')  
    
    if value.strip() == '':  
        return None
    if value.strip() == ' ':  

        return None

    return str(value.strip())

columns = ["Month", "Name", "Type_of_Loan", "Payment_Behaviour",]
data[columns] = data[columns].applymap(clean_text)

data.head()
```

```python
import re

def convert_month(age_str):
    try:
        match = re.search(r'(\d+)\s*Years?.*?(\d+)\s*Month', age_str)
        if match:
            years = int(match.group(1))
            months = int(match.group(2))
            return years * 12 + months
        match_years = re.search(r'(\d+)\s*years?', age_str)
        if match_years:
            return int(match_years.group(1)) * 12
        return 0  
    except:
        return 0  
data['Credit_Age_Month'] = data['Credit_History_Age'].apply(convert_month)
```

```python
data = data.drop(["Credit_History_Age"], axis=True)
```

```python
data["Age"] = data["Age"].apply(lambda x: x if 0 <= x <= 120 else np.nan)
data["Annual_Income"] = data["Annual_Income"].apply(lambda x: x if 0 <= x  else np.nan)
data["Monthly_Inhand_Salary"] = data["Monthly_Inhand_Salary"].apply(lambda x: x if 0 <= x  else np.nan)
data["Num_Bank_Accounts"] = data["Num_Bank_Accounts"].apply(lambda x: x if 0 <= x  else np.nan)
data["Num_Credit_Card"] = data["Num_Credit_Card"].apply(lambda x: x if 0 <= x  else np.nan)
data["Interest_Rate"] = data["Interest_Rate"].apply(lambda x: x if 0 <= x  else np.nan)
data["Num_of_Loan"] = data["Num_of_Loan"].apply(lambda x: x if 0 <= x  else np.nan)
data["Delay_from_due_date"] = data["Delay_from_due_date"].apply(lambda x: x if 0 <= x  else np.nan)
data["Num_of_Delayed_Payment"] = data["Num_of_Delayed_Payment"].apply(lambda x: x if 0 <= x  else np.nan)
data["Num_Credit_Inquiries"] = data["Num_Credit_Inquiries"].apply(lambda x: x if 0 <= x  else np.nan)
data["Outstanding_Debt"] = data["Outstanding_Debt"].apply(lambda x: x if 0 <= x  else np.nan)
data["Outstanding_Debt"] = data["Outstanding_Debt"].apply(lambda x: x if 0 <= x  else np.nan)
```
